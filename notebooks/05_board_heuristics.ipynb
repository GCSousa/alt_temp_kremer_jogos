## 1. O Tabuleiro do Mu Torere

Mu Torere possui 8 casas externas (Kewai) circundando 1 casa central (Putahi). 
Há 8 peças no total (4 de cada jogador) e 1 espaço vazio.

- Uma peça externa só pode mover para a casa central se estiver adjacente a uma peça adversária (para evitar empates rápidos e bloqueios imediatos).
- Quem não conseguir mover na sua vez perde a partida.

## 2. Analisando Estados e Heurísticas de Mobilidade

Vamos avaliar diferentes configurações do tabuleiro para ver como a função `evaluate` quantifica a vantagem posicional de cada jogador com base no número de movimentos disponíveis.

```python
game = MuTorereV1()

# Caso A: Estado inicial equilibrado
state_a = game.get_initial_state()

# Caso B: P1 (X) muito próximo de vencer, com P2 (O) sem movimentos
state_b = {
    "board": [
        1, 1, 1, 1,   # Kewai 0-3
        2, 2, 2, 2,   # Kewai 4-7
        0             # Putahi (centro) vazio
    ],
    "current_player": 2 # Vez do jogador 2
}

def analyze_heuristic(state_dict, label):
    board = state_dict["board"]
    current = state_dict["current_player"]
    
    # Calcular mobilidade (número de movimentos válidos)
    valid_moves = game.get_valid_moves(state_dict)
    score = game.evaluate(state_dict)
    result = game.check_result(state_dict)
    
    print(f"=== Cenário: {label} ===")
    print(f"  Tabuleiro: {board}")
    print(f"  Vez do Jogador: {current}")
    print(f"  Movimentos válidos possíveis: {len(valid_moves)}")
    print(f"  Pontuação Heurística: {score}")
    print(f"  Resultado imediato: {result}")
    print()

analyze_heuristic(state_a, "Início equilibrado")
analyze_heuristic(state_b, "P2 prestes a ser bloqueado (P1 vence)")
```

## 3. Conclusão sobre as Heurísticas de Bloqueio

Como o Mu Torere é um jogo de bloqueio puro, a heurística de **Mobilidade Relativa** (diferença entre movimentos válidos do P1 e P2) é extremamente eficiente:

- Pontuação positiva alta: Ampla vantagem posicional e maior mobilidade para o Jogador 1.
- Pontuação negativa alta: Jogador 2 possui controle espacial superior, encurralando o Jogador 1.
- Pontuação zero: Equilíbrio absoluto ou empate técnico.
